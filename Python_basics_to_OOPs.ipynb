```python
def prime(n):
  c=0
  for i in range(2,n):
    if n%i==0:
      print(f"{n} is not a prime number")
      c+=1
      break
  if c==0:
    print(f"{n} is prime number")
```

```python
class buisiness():
  def bonus(self,name,age,salary):
    self.name=name
    self.age= age
    self.salary = salary
  def calculate(self):
    self.bonus = self.salary/100*10
    self.salary = self.salary+self.bonus
  def salary_bonus(self):
    return (f"name : {self.name} age : {self.age} bonus: {self.bonus} Totalsalary : {self.salary} ")
obj= buisiness()
obj.bonus("siddhi",26,56000)
obj.calculate()
obj.salary_bonus()
```

```python
class expensess():
  def __init__(self,sal_1,sal_2):
    self.__Father =sal_1
    self.__mother = sal_2
    self.__Total_salary= self.__Father + self.__mother
  def Income(self,Amt):
    self.__Total_salary +=Amt
  def Exp(self,Amt):
    self.__Total_salary -= Amt
  def Display_Total(self):
    return self.__Total_salary
```
