```python
class Employee():
  empcount=0
  incmt =1.10
  def __init__(self,first,last,salary):
    self.first= first
    self.last = last
    self.salary= salary
    self.email = first.lower()+'.'+last.lower()+'@gmail.com'
    Employee.empcount+=1
  def full_name(self):
    self.full_name1 = self.first.title()+' '+self.last.title()
    return self.full_name1
  def increment(self):
    self.salary = self.salary*Employee.incmt
    return round(self.salary)
```

```python
emp1 = Employee('vedika','goverkar',55000)
emp2= Employee('saniya','shaikh',45000)
emp3= Employee('swara','shingate',32000)
emp4 = Employee('sanjay','humar',25000)
emp5 = Employee('Harish','patil',25000)
```

```python
class manager(Employee):
  def __init__(self,first,last,salary,employees=None):
    super().__init__(first,last,salary)

    if employees is None:
      self.employees=[]
    else:
      self.employees = employees
  def addemp(self,emp):
    if emp not in self.employees:
        self.employees.append(emp)
  def removeemp(self,emp):
    if emp in self.employees:
        self.employees.remove(emp)
  def printemp(self):
    for emp in self.employees:
      print('--->',emp.full_name())
```
