```python
#17. Write an algorithm/ program to print the factorial of number accepted from user.
a = int(input("enter a non-negative number "))
factorial = 1
if a < 0:
  print("factorial does not exists for this number")
elif a == 0:
  print("factorial of 0 is 1")
else:
  for i in range(1,a+1):
    factorial *= i
  print(f"The factorial of {a} is {factorial}")
```
